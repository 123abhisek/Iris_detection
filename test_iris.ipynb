```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.image import imread
import tensorflow as tf
from keras.preprocessing import image
```

```python
img_height=180
img_width=180

model = tf.keras.models.load_model('Iris Flask Project/model/Iris_Model.h5')
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('Iris Flask Project/model/data_set.csv', encoding='latin-1')
df = df.replace('',np.nan)
df = df.dropna(axis="columns",how="any")
```

```python
path = "Dataset/test/iris_virginica/iris-e0022650ad7be128f517b70d30b0462d2a69bae0202ad53c3cac1c4894f19ef9.jpg"
```

```python
img = image.load_img(path,target_size=(img_height,img_width))
img = image.img_to_array(img)
img = np.expand_dims(img,axis=0)
# img = preprocess_input(img)


class_names = ['iris_setosa', 'iris_versicolour', 'iris_virginica']

prediction = model.predict(img)

predicted_class = np.argmax(prediction)

print(f"Predicted class : {predicted_class}")
print(f"Predict Disease : {class_names[predicted_class]}")



# Assuming df is a DataFrame with appropriate columns
# Class_Name, Varieties, Facts, Uses, Petals, Sepals, Flowers, Leaves, Height, Habitat = df.loc[predicted_class, ['Class Name', 'Varieties', 'Facts', 'Uses', 'Petals', 'Sepals', 'Flowers', 'Leaves', 'Height', 'Habitat']]

# img = imread(path)
# plt.imshow(img)
# plt.axis('off')
# plt.show()

# print(f"\nClass_Name : {Class_Name}\nVarieties : {Varieties}\nFacts : {Facts}\nUses : {Uses}\nPetals : {Petals}\nSepals : {Sepals}\nFlowers : {Flowers}\nLeaves : {Leaves}\nHeight : {Height}\nHabitat : {Habitat}")
```
